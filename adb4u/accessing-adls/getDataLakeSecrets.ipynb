#### How to use [Secrets?](https://docs.microsoft.com/en-us/azure/databricks/security/secrets/)

```
dbutils.help()
```

```
dbutils.secrets.help()
```

```
%scala
lazy val datalakename = "datalakename"
lazy val containername = "container"
lazy val datalake_key = dbutils.secrets.get("scopeName","lakeAccessKey")
lazy val sp_client_id = dbutils.secrets.get("scopeName","clientId") 
lazy val sp_client_key = dbutils.secrets.get("scopeName","clientSecret")
lazy val tenant_id = dbutils.secrets.get("scopeName","azureTenantId")


spark.conf.set(s"fs.azure.account.auth.type.$datalakename.dfs.core.windows.net", "OAuth")
spark.conf.set(s"fs.azure.account.oauth.provider.type.$datalakename.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set(s"fs.azure.account.oauth2.client.id.$datalakename.dfs.core.windows.net", sp_client_id)
spark.conf.set(s"fs.azure.account.oauth2.client.secret.$datalakename.dfs.core.windows.net",sp_client_id)
spark.conf.set(s"fs.azure.account.oauth2.client.endpoint.$datalakename.dfs.core.windows.net", s"https://login.microsoftonline.com/$tenant_id/oauth2/token")


spark.sparkContext.hadoopConfiguration.set("fs.azure.account.auth.type.$datalakename.dfs.core.windows.net", "OAuth")
spark.sparkContext.hadoopConfiguration.set("fs.azure.account.oauth.provider.type.$datalakename.dfs.core.windows.net",  "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.sparkContext.hadoopConfiguration.set("fs.azure.account.oauth2.client.id.$datalakename.dfs.core.windows.net", sp_client_id)
spark.sparkContext.hadoopConfiguration.set("fs.azure.account.oauth2.client.secret.$datalakename.dfs.core.windows.net", sp_client_key)
spark.sparkContext.hadoopConfiguration.set("fs.azure.account.oauth2.client.endpoint.$datalakename.dfs.core.windows.net", s"https://login.microsoftonline.com/$tenant_id/oauth2/token")

print(s"Spark configs set for: $datalakename")
dbutils.fs.ls(s"abfss://$containername@labsdatalake.dfs.core.windows.net/")
```
